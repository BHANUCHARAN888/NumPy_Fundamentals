```python
import numpy as np
```

```python
# From List
print("1D Array")
print(np.array([1,2,3,4,5])) # 1D
print("-----------------------")
print("2D Array")
print(np.array([[1,2,3],[4,5,6]])) # 2D
```

```python
# From Scratch
print("zeros")
print(np.zeros((3,3)))
print("-----------------------")
print("ones")
print(np.ones((3,3)))
print("-----------------------")
print("full")
print(np.full((2,3),20))
print("-----------------------")
print("eye")
print(np.eye(4))
print("-----------------------")
print("arange")
print(np.arange(1,11))
print(np.arange(1,11,5))
print("-----------------------")
print("linspace")
print(np.linspace(0,1,5))
print("-----------------------")
```

```python
# Checking array Properties
arr = np.array([[1,2,3,4],[4,5,6,7]])
print(arr)
print("-----------------------")
print("Shape: ",arr.shape)
print("Size: ",arr.size)
print("Dimensions: ",arr.ndim)
print("Data type: ",arr.dtype)
```

```python
# Changing DataTypes
arr = np.array([1,2,3], dtype=np.float64)
print(arr.dtype)
print(arr)

arr_int = arr.astype(np.int64)
print(arr_int)
```

```python
# Reshaping and Flattening array
arr = np.array([[1, 2, 3], [4, 5, 6]])
print(arr.shape)
print("-------------------------------")
reshaped = arr.reshape((3, 2))
print(reshaped)
print("-------------------------------")
flattened = arr.flatten()  # Convert 2D → 1D
print(flattened)
```
