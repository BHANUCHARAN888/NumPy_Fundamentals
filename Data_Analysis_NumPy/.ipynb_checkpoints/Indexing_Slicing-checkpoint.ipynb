```python
import numpy as np
arr = np.array([10, 20, 30, 40])
```

```python
# indexing
print(arr[0])  # 10
print(arr[-1]) # 40
```

```python
# slicing
print(arr[0:4])
print(arr[::2])
print(arr[::-1])
```

### ** Slicing returns a view, not a copy! Changes affect the original array.**
#### Why?
- Memory Efficiency: Avoids unnecessary copies, making operations faster and saving memory.
- Performance: Enables faster access and manipulation of large datasets without duplicating data.

```python
arr = np.array([1,3,5,7,9,11,13,15])
print("Before Slicing: ",arr)
sliced = arr[1:6]
sliced[0] = 888
print("Slicing part: ",sliced)
print("After Slicing: ",arr)
```

```python
sliced = sliced[1:6].copy()
sliced[0] = 999
print("Sliced part return copy not original: ",sliced)
print("Original not changed because of copy funtion: ",arr)
```

### Fancy Indexing
- Return the selected indices
- arr[index1, undex2,..]

```python
arr = np.array([20,30,40,50,60,70,80])
index = [1,3,5]
print(arr[index])
```

### Boolean Masking
- This is a powerful way to filter large datasets efficiently!

```python
arr = np.array([10,4,27,2,8,19,11,3,22,1])
print(arr>20)
print(arr[arr>20])
```
