```python
import numpy as np
import time
```

## List
- store elements as pointers, instead of continuous block of memory
- lack of vectorization
- consume large storage

```python
# List
size = 1_000_000
l1 = list(range(size))
l2 = list(range(size))

start = time.time()
result = [x+y for x,y in zip(l1,l2)]
end = time.time()

# print(result) throws an error
print(end - start)
```

## NumPy
- Far better than Python List
- vectorized operations(no loop complexity)
- uses less memory
- has built-in mathematical functions

```python
# Numpy
size = 1_000_000
arr1 = np.array(l1)
arr2 = np.array(l2)

start = time.time()
result = arr1 + arr2
end = time.time()

print(result)
print(end - start)
```

```python
# More Efficiency
import sys
list_data = list(range(1000))
numpy_data = np.array(list_data)

print("List_size:",sys.getsizeof(list_data)*len(list_data)," bytes")
print("Numy_size:",numpy_data.nbytes," bytes")
```

```python
# Vectorization - SIMD(single instruction multiple data) CPU level optimation

# List
list1 = [1,2,3,4,5,6]
list_squares = [x**2 for x in list1] # loop based
print(list_squares)

# Numpy
arr1 = np.array(list1)
numpy_square = arr1**2 # vectorized
print(numpy_square)
```
